# Generating ramachandran plots from Alpha Fold Predicted Structures

```python
from Bio.PDB.MMCIFParser import MMCIFParser
from Bio.PDB.Polypeptide import PPBuilder
```

```python
green_and_yellow_prot_CIF = RvsS_df[RvsS_df['label'].isin(['green', 'yellow'])]['Protein ID'].to_list()
```

```python
green_and_yellow_prot_CIF
```

```python
def extract_phi_psi_angles(cif_file):
    parser = MMCIFParser()
    structure = parser.get_structure('structure', cif_file)
    ppb = PPBuilder()
    phi_psi_aa_position = []
    for model in structure:
        for chain in model:
            for pp in ppb.build_peptides(chain):
                residues = pp.get_residues()
                phi_psi = pp.get_phi_psi_list()
                for i, (residue, (phi, psi)) in enumerate(zip(residues, phi_psi)):
                    if phi is not None and psi is not None:
                        residue_id = residue.get_id()[1]
                        amino_acid = residue.get_resname()
                        phi_psi_aa_position.append((amino_acid, residue_id, phi, psi))
    return phi_psi_aa_position

# Directory containing CIF files
cif_directory = '../alphafold_data/cif'

# Dictionary to store phi, psi angles, amino acid, and position for each protein
protein_dataframes = {}

# Iterate over CIF files
for id in green_and_yellow_prot_CIF:
    cif_file = id + '.cif'
    if cif_file.endswith('.cif'):
        protein_id = os.path.splitext(cif_file)[0]
        cif_path = os.path.join(cif_directory, cif_file)
        data = extract_phi_psi_angles(cif_path)
        df = pd.DataFrame(data, columns=['Amino Acid', 'Position', 'Phi', 'Psi'])
        protein_dataframes[protein_id] = df

# Access DataFrame for a specific protein
protein_id = 'Q8C196'
print(protein_dataframes[protein_id])
```
