# Below is a visualization of Methionines in the first RvsS dataset

```python
RvsS_df = pd.read_csv('../RvsS/RvsS_peptides_completed_sequence_with_thresholds.csv')
```

```python
RvsS_df
```

```python
dihedral_df = pd.read_csv('dihedral_angles.csv')
```

```python
dihedral_df
```

```python
green_peptides = RvsS_df[RvsS_df['label'] == 'green']
green_peptides
```

```python
green_peptides.shape
```

```python
green_PHI = []
green_PSI = []
for i in range(green_peptides.shape[0]):
    protein = green_peptides.iloc[i]['Protein ID'] 
    aa_position = green_peptides.iloc[i]['Methionine Location'] + 1
    row_of_interest = dihedral_df.loc[(dihedral_df['Protein ID'] == protein) & (dihedral_df['Residue Position'] == aa_position)]

    if row_of_interest['Residue Name'].iloc[0] == 'MET':
        green_PHI.append(row_of_interest['PHI degrees'].iloc[0])
        green_PSI.append(row_of_interest['PSI degrees'].iloc[0])
    else:
        print(protein, row_of_interest['Residue Name'].iloc[0])
```

```python
green_PHI
```

```python
len(green_PSI)
```

```python
yellow_peptides = RvsS_df[RvsS_df['label'] == 'yellow']
yellow_peptides
```

```python
yellow_PHI = []
yellow_PSI = []
for i in range(yellow_peptides.shape[0]):
    protein = yellow_peptides.iloc[i]['Protein ID'] 
    aa_position = yellow_peptides.iloc[i]['Methionine Location'] + 1
    row_of_interest = dihedral_df.loc[(dihedral_df['Protein ID'] == protein) & (dihedral_df['Residue Position'] == aa_position)]

    if row_of_interest['Residue Name'].iloc[0] == 'MET':
        yellow_PHI.append(row_of_interest['PHI degrees'].iloc[0])
        yellow_PSI.append(row_of_interest['PSI degrees'].iloc[0])
    else:
        print(protein, row_of_interest['Residue Name'].iloc[0])
```

```python
plt.figure(figsize=(10, 8))
# sns.kdeplot(x=green_PHI, y=green_PSI, fill=False, bw_adjust=0.5, label = 'green', color='green')
# sns.kdeplot(x=yellow_PHI, y=yellow_PSI, fill=False, bw_adjust=0.5, label='yellow', color = 'yellow')

sns.scatterplot(x=green_PHI, y=green_PSI, color='blue')
sns.scatterplot(x=yellow_PHI, y=yellow_PSI, color = 'red')

plt.title('Ramachandran Plot')
plt.xlabel('PHI')
plt.ylabel('PSI')
plt.xlim(-180, 180)
plt.ylim(-180, 180)

plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='black', linestyle='--', linewidth=1)

plt.legend()

plt.show()
```

# MsrAKD visualization

```python
MsrAKD_df = pd.read_csv('../MsrKD/MsrAKD_with_alphafold.csv')
```

```python
MsrAKD_df
```

```python
colors = MsrAKD_df['Color'].unique()
colors
```

```python
plt.figure(figsize=(10, 8))
plt.figure(dpi=1200)
for color in colors:
    if color == 'white':
        continue
    PHI_lst = []
    PSI_lst = []
    sub_df = MsrAKD_df[MsrAKD_df['Color'] == color]
    for i in range(sub_df.shape[0]):
        protein = sub_df.iloc[i]['Protein ID'] 
        aa_position = sub_df.iloc[i]['Methionine Location'] + 1
        row_of_interest = dihedral_df.loc[(dihedral_df['Protein ID'] == protein) & (dihedral_df['Residue Position'] == aa_position)]
        if row_of_interest.shape[0] == 0:
            print(protein, aa_position)
            continue

        if row_of_interest['Residue Name'].iloc[0] == 'MET':
            PHI_lst.append(row_of_interest['PHI degrees'].iloc[0])
            PSI_lst.append(row_of_interest['PSI degrees'].iloc[0])
        else:
            print(protein, row_of_interest['Residue Name'].iloc[0], aa_position)
            continue
    sns.scatterplot(x=PHI_lst, y=PSI_lst, color=color, s = 10)
    sns.scatterplot(x=PHI_lst, y=PSI_lst, color = color, s = 10)
plt.title('Ramachandran Plot')
plt.xlabel('PHI')
plt.ylabel('PSI')
plt.xlim(-180, 180)
plt.ylim(-180, 180)

plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='black', linestyle='--', linewidth=1)

plt.legend()

plt.show()
    
    
```
