```python
extracted_col = ['Modified Sequence', 'Gene', 'Protein Description']
extracted = df[extracted_col][(df['Modified Sequence'].astype(str).str.contains('655.3735')) | (df['Modified Sequence'].astype(str).str.contains('649.3660'))]
extracted.head(5)
```

```python
extracted.shape
```

```python
extracted['Modified Sequence'].values[0].split('[655.3735]')
re.split(r'\[649.3660]|\[655.3735]', extracted['Modified Sequence'].values[1])
```

```python
sequence = []

for seq in extracted['Modified Sequence'].values:
    sequence.append(re.split(r'\[649.3660]|\[655.3735]', seq))

sequence[:5]
```

```python
left = []
right = []

for elem in sequence:
    left.append(elem[0][:-1])
    right.append(elem[1])

for i in range(len(left)):
    left[i] = left[i].replace('n[42.0106]', '')

print(left[-5:])
print(right[:5])
```

```python
left[0][:2]
```

```python
for i in range(len(left)):
    left[i] = left[i][-10:]
    right[i] = right[i][:10]

print(left)
print(right)
```

```python
extracted['Modified Sequence'].values[0].replace('M[655.3735]', '').replace('M[649.3660]', '').replace('n[42.0106]', '')
```

```python
extracted
```

```python
extracted['Pre-Sequence'] = left
extracted['Post-Sequence'] = right
extracted['Cleaned Sequence'] = extracted['Modified Sequence'].apply(lambda row: row.replace('M[655.3735]', '').replace('M[649.3660]', '').replace('n[42.0106]', ''))

extracted = extracted[['Gene', 'Modified Sequence', 'Cleaned Sequence', 'Pre-Sequence', 'Post-Sequence', 'Protein Description']]
extracted.head(5)
```

```python
def dataframe_to_fasta(df, output_file):
    with open(output_file, 'w') as f:
        for index, row in df.iterrows():
            f.write(f'>{row["Gene"]}\n{row["Cleaned Sequence"]}\n')

dataframe_to_fasta(extracted[extracted['Cleaned Sequence'].apply(len) == 9], 'output.fasta')
```
