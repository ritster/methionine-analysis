```python
import pandas as pd
```

```python
path = '../datasets/'
df = pd.read_csv(path + 'inaccessible_methionines_dataset.csv')

df.head(5)
```

```python
len(df['Right 10'].values) == 10
```

```python
left = df['Left 10'].values 
right = df['Right 10'].values

res = []

for l, r in zip(left, right):
    if len(str(l)) == 10 and len(str(r)) == 10:
        res.append(str(l) + 'M' + str(r))

print(res)
print(len(res))
```

```python
file_path = 'inaccessible_M.txt'

# Open the file in write mode
with open(file_path, 'w') as file:
    # Write each element of the array to the file
    for item in res:
        file.write(item + '\n')

print(f"Array has been written to {file_path}")
```
