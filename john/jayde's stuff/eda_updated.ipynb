```python
complete_seq = []

for i in range(len(df['Complete Sequence'])):
    start = df['Start'][i] -1 
    end = df['End'][i]
    m = df['Complete Sequence'][i][start:end].find('M')
    complete_seq.append(df['Complete Sequence'][i][start+m-10:start+m+11])
```

```python
complete_seq[1], len(complete_seq[1])
```

```python
counter = 0
false_seq = []
selected = []
for value in complete_seq:
    length = len(value)
    print(length == 21)
    if length == 21:
        selected.append(value)
    if length != 21:
        print(value)
        false_seq.append(value)
        counter +=1
```

```python
false_seq[:5]
```

```python
len(selected) == (734 - len(false_seq))
```

```python
df['Complete Sequence (10 each side)'] = complete_seq
df.head(5)
```

```python
filtered_df = df[df['Complete Sequence (10 each side)'].str.len() == 21]
```

```python
len(filtered_df), len(selected)
```

```python
# Create a folder named "datasets" if it doesn't exist
folder_name = "datasets"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Define the file path
file_path = os.path.join(folder_name, "proteins_10_each_side.csv")

# Export the DataFrame to a CSV file
df.to_csv(file_path, index=False)

print(f"DataFrame exported to: {file_path}")
```
