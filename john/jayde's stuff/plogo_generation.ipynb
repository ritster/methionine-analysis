```python
df = df.dropna(subset=['Right 20', 'Left 20'])
```

```python
def generate(color):
    left = color['Left 20'].values
    right = color['Right 20'].values

    left_5, left_10, left_15, left_20 = [], [], [], []

    for left_protein in left:
        left_5.append(left_protein[-5:])
        left_10.append(left_protein[-10:])
        left_15.append(left_protein[-15:])
        left_20.append(left_protein)

    right_5, right_10, right_15, right_20 = [], [], [], []

    for right_protein in right:
        right_5.append(right_protein[:5])
        right_10.append(right_protein[:10])
        right_15.append(right_protein[:15])
        right_20.append(right_protein)

    pro_5, pro_10, pro_15, pro_20 = [], [], [], []

    for i in range(len(left)):
        mod = 'M'
        pro_5.append(left_5[i] + mod + right_5[i])
        pro_10.append(left_10[i] + mod + right_10[i])
        pro_15.append(left_15[i] + mod + right_15[i])
        pro_20.append(left_20[i] + mod + right_20[i])
    
    pro_5t, pro_10t, pro_15t, pro_20t = [], [], [], []

    for i in range(len(left)):
        if len(pro_5[i]) == 11:
            pro_5t.append(pro_5[i])
        if len(pro_10[i]) == 21:
            pro_10t.append(pro_10[i])
        if len(pro_15[i]) == 31:
            pro_15t.append(pro_15[i])
        if len(pro_20[i]) == 41:
            pro_20t.append(pro_20[i])
        
    return pro_5t, pro_10t, pro_15t, pro_20t
```

```python
def create_txt(list, names):
    t5, t10, t15, t20 = list[0], list[1], list[2], list[3]
    name_5t, name_10t, name_15t, name_20t = names[0], names[1], names[2], names[3]

    with open(name_5t + '.txt', 'w') as f:
    # Iterate over the array elements
        for item in t5:
            # Write each element to the file
            f.write(str(item) + '\n')

    with open(name_10t + '.txt', 'w') as f:
    # Iterate over the array elements
        for item in t10:
            # Write each element to the file
            f.write(str(item) + '\n')

    with open(name_15t + '.txt', 'w') as f:
    # Iterate over the array elements
        for item in t15:
            # Write each element to the file
            f.write(str(item) + '\n')
    
    with open(name_20t + '.txt', 'w') as f:
    # Iterate over the array elements
        for item in t20:
            # Write each element to the file
            f.write(str(item) + '\n')
```

```python
reactive = df[df.prelim_hyperreactive == True]
non_reactive = df[df.prelim_hyperreactive == False]

print(reactive.shape, non_reactive.shape)
```

```python
reactive_5, reactive_10, reactive_15, reactive_20 = generate(reactive)
non_reactive_5, non_reactive_10, non_reactive_15, non_reactive_20 = generate(non_reactive)
```

```python
reactive_list = [reactive_5, reactive_10, reactive_15, reactive_20]
non_reactive_list = [non_reactive_5, non_reactive_10, non_reactive_15, non_reactive_20]

reactive_names = ['reactive_5', 'reactive_10', 'reactive_15', 'reactive_20']
non_reactive_names = ['non_reactive_5', 'non_reactive_10', 'non_reactive_15', 'non_reactive_20']
```

```python
create_txt(reactive_list, reactive_names)
```

```python
create_txt(non_reactive_list, non_reactive_names)
```
